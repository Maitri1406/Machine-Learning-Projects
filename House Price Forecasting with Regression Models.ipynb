# Part-1: data Exploration and Pre-processing

```python
# Drop address, date, postcode, YearBuilt, lattitude, longtitude columns 
df=df.drop(['Address', 'Date','Postcode','YearBuilt','Lattitude','Longtitude'],axis=1)
```

```python
df
```

```python
#  Find the count of null value in each column 
df.isnull().sum()
```

```python
# Fill the null value of property count, distance, Bedroom2, Bathroom, Car with 0
df[['Distance', 'Bedroom2', 'Bathroom', 'Car','Propertycount']]=df[['Distance', 'Bedroom2', 'Bathroom', 'Car','Propertycount']].fillna(0)
```

```python
# Fill Null value of land size and bidding area columns with Mean 
df['Landsize']=df['Landsize'].fillna(df.Landsize.mean())
df['BuildingArea']=df['BuildingArea'].fillna(df.BuildingArea.mean())
```

```python
df.isnull().sum()
```

```python
# drop all null column
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
# find the unique value in method column 
df['Method'].unique()
```

```python
# Create a dummy data for categorical data 
df=pd.get_dummies(df,drop_first=True)
df.shape
```

# Part-2: Working with Model 

```python
#  Create the target data and feature data where target data is price 
x=df.drop('Price',axis=1)
y=df['Price']
```

```python
x
```

```python
y=pd.DataFrame(y)
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
# Create a linear regression model for Target and feature data 
reg=LinearRegression()
reg.fit(x_train,y_train)
```

```python
# Check if the model is overfitting or underfitting or it is accurate 
reg.score(x_train,y_train)
```

```python
reg.score(x_test,y_test)
# this model is overfetting
```

# Ridge

```python
# If the model is overfitting then apply ridge and lasso regression algorithms
ridge_reg=Ridge(alpha=50,max_iter=100,)
ridge_reg.fit(x_train,y_train)
```

```python
ridge_reg.score(x_test,y_test)
```

```python
ridge_reg.score(x_test,y_test)
```

```python
# Extract slope and intercept value from the model 
slope=reg.coef_
Intercept=reg.intercept_
print("Slope(coefficient):",slope)
print("Intercept:",Intercept)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
prediction=reg.predict(x_test)
```

```python
# Display Mean Squared Error 
MSE=mean_squared_error(y_test,prediction)
print("Mean Squared Error:",MSE)
```

```python
# Display Mean Absolute Error 
MAE=mean_absolute_error(y_test,prediction)
print("Mean Absolute Error:",MAE)
```

```python
# Display Root mean Squared error 
RMSE=np.sqrt(MSE)
print("Root Mean Squeared Error:",RMSE)
```

```python
# Display R2 score 
R2=r2_score(y_test,prediction)
print("R2 Score:",R2)
```
